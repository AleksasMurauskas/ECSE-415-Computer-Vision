```
#List to hold Image vectors 
train_vector =[]
test_vector=[]
tot_data =[]
img_size = new_h*new_w

tot_data = complete_test+complete_train

#Create a new lest  to act as the vector 
for active in complete_train:
  f1_face=active.reshape(img_size,)
  train_vector.append(f1_face)
active_vector =np.asarray(train_vector)
active_vector= active_vector.transpose()

#Creation of a test vector
for active in complete_test: 
  face = active.reshape(img_size,)
  test_vector.append(face)

test_vector = np.asarray(test_vector)
test_vector= test_vector.transpose()

#print("Vector shape check : "+str(active_vector.shape))
#Find Mean to Normalize Vector
#FInd the Mean along the axis
mean_vector = active_vector.mean(axis=1)
#Stretch the mean to the shape of the original vector. 
mean_vector_reshaped = mean_vector.reshape(active_vector.shape[0],1)
#Normalize the vector by subtracting the found mean 
normalized_vector = active_vector -mean_vector_reshaped
#Find Covariance
trans_vector = np.transpose(normalized_vector)
cov = np.cov(trans_vector)
#Now to find EigenVectors 
eig_vals, eig_vectors = np.linalg.eig(cov)
#print(eig_vals)
#print(eig_vectors.shape)

#Repeat for the test cases
mean_test_vector = test_vector.mean(axis=1)
mean_test_vector_re =  mean_test_vector.reshape(test_vector.shape[0],1)
test_norm_vector = test_vector - mean_test_vector_re
test_trans_vector = np.transpose(test_norm_vector)
cov2 = np.cov(test_trans_vector)
_, test_eig = np.linalg.eig(cov2)
#print(test_eig.shape)

#Testing with eigenface over whole dataset
data_set=[]
for item in tot_data:
  f1_face=item.reshape(img_size,)
  data_set.append(f1_face)
data_vector = np.asarray(data_set)
data_vector= data_vector.transpose()
tot_mean = data_vector.mean(axis=1)
tot_mean_re = tot_mean.reshape(data_vector.shape[0],1)
norm_set = data_vector- tot_mean_re
trans_data = np.transpose(norm_set)
cov3 = np.cov(trans_data)
_,total_eig_vector  = np.linalg.eig(cov3)
```
